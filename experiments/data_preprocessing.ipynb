```python
import pandas as pd 

df = pd.read_csv("dataset.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
movie_ids = {movie: idx + 1 for idx, movie in enumerate(df['Movie_Name'].unique())}
```

```python
print(movie_ids)
```

```python
df['Movie_Id'] = df['Movie_Name'].apply(lambda x: movie_ids[x])
```
