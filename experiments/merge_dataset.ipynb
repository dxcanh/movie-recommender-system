```python
import pandas as pd

# read data from csv files
ratings = pd.read_csv('/home/canhdx/workspace/movie-recommender-system/data/ratings.csv')
movies = pd.read_csv('/home/canhdx/workspace/movie-recommender-system/data/movies.csv')

# get year from movie name
movies['Year'] = movies['title'].str.extract(r'\((\d{4})\)')
movies['Movie_Name'] = movies['title'].str.replace(r'\(\d{4}\)', '', regex=True).str.strip()

# merge two dataframes
merged_df = pd.merge(ratings, movies, on='movieId', how='inner')

# sort by userId and rating
final_df = merged_df[['userId', 'Movie_Name', 'rating', 'genres', 'Year', 'movieId']]

print(final_df.head())
```

```python
final_df = final_df.rename(columns={'userId': 'User_Id', 'Movie_Name': 'Movie_Name', 'rating': 'Rating', 'genres': 'Genre', 'Year': 'Year', 'movieId': 'Movie_Id'})
```
